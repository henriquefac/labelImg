```python
from configPy import Config
from pathlib import Path
from PIL import Image
from src.aux_imgs import *
import numpy as np
import shutil
```

```python
file_dir = Config.get_dir_files()

# new data path
new_data_dir = file_dir["new_data"]

# data path (dados originais)
data_dir = file_dir["data"]
```

```python

# lista com o path para as imagens do new_data
lista_new_data_path:list[Path] = list(new_data_dir.list_files().values())
# dir com as labels
labels_new_data_dir = new_data_dir["newData_Label"]

new_data_img_labels = [(img_path, labels_new_data_dir.by_name(img_path.stem)[0]) for img_path in lista_new_data_path if len(labels_new_data_dir.by_name(img_path.stem)) > 0]
```

```python
# diretório das imagens dentro de data
img_data_list: list[Path] = list(data_dir["imgs"].list_files().values())
label_data_dir = data_dir["data_label"]

data_img_labels = [(img_path, label_data_dir.by_name(img_path.stem)[0]) for img_path in img_data_list if len(label_data_dir.by_name(img_path.stem)) > 0]
```

```python
print(data_img_labels)
```

```python
# new dir merge_data
merge_data = file_dir.create_dir("merge_dir")

# sub direttórios
imgs_merge = merge_data.create_dir("imgs")
labels_merge = merge_data.create_dir("labels")

data_img_labels.extend(new_data_img_labels)

for i, tupla_img_label in enumerate(data_img_labels):
    img, label = tupla_img_label

    new_img = imgs_merge.create_file_path(str(i+1), img.suffix[1:])
    new_label = labels_merge.create_file_path(str(i+1), label.suffix[1:])

    try:
        shutil.copy(img, new_img)
        print(f"Copiado: {img.name} para {new_img.name}")
        shutil.copy(label, new_label)
        print(f"Copiado: {label.name} para {new_label.name}")
    except FileNotFoundError:
        print(f"Erro: Arquivo não encontrado ao tentar copiar {img} ou {label}.")
    except Exception as e:
        print(f"Ocorreu um erro ao copiar os arquivos: {e}")
```

```python
# alterar classe das labels que deveriam ser 0 (filho da PUUUUUUUUUUUUUUTAAAAAAAAAA)

lista_de_labels: list[Path] = list(labels_merge.list_files().values())

for file in lista_de_labels:
    with open(file, "r") as f:
        prime_id = int(f.readline().split()[0])
        
    if prime_id == 0:
        continue
    print(prime_id)
    with open(file, "r") as f:
        lines = f.readlines()

    for i, line in enumerate(lines):
        split_line = line.split()
        split_line[0] = "0"
        lines[i] = " ".join(split_line)
    
    with open(file, "w") as f:
        f.write("\n".join(lines))
        
```

- Falsos negativos: 225. O modelo deixou passar 225 "Waldos" sem serem vistos.

### Cálculos
- Cálculo do Recall (Acertos / (Falsos negativos + Acertos)): 272 / (225 + 272) = 272 / 497 ≈ 54.7%
- Cálculo da Precisão (Acertos / (Acertos + Falsos positivos)): 272 / (272 + 61) = 272 / 333 ≈ 81.7%

A Curva de Precisão-Confiança demonstra a relação direta entre a confiança do modelo em suas previsões e a precisão real dessas previsões. O gráfico revela uma tendência positiva e acentuada, indicando que quanto mais confiante o modelo está em uma detecção, maior a probabilidade de ela estar correta.
O ponto mais relevante é destacado na legenda: o modelo alcança 100% de precisão (1.00) quando o limiar de confiança é ajustado para 0.912.
Em resumo, o gráfico confirma que o modelo é capaz de gerar detecções extremamente confiáveis. Ao configurar a aplicação para aceitar apenas previsões com um alto nível de confiança (acima de 91.2%), é possível eliminar virtualmente todos os falsos positivos, garantindo que apenas os resultados mais precisos sejam considerados.
